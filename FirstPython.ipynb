```
#-------------------------------------Slicing-------------------------------------------------------#

#Slicing means taking elements from one given index to another given index

#Eg: [start:end]
#Eg: [start:end:step]

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])


print("Slice elements from index 1 to index 5 from the following array");
print(arr[1:5]);
print();

print("Slice elements from index 4 to the end of the array");
print(arr[4:]);
print();

print("Slice elements from the beginning to index 4 (not included)");
print(arr[:4]);
print();

print("Slice from the index 3 from the end to index 1 from the end");
print(arr[-3:-1]);
print();

print("Return every other element from index 1 to index 5");
print(arr[1:5:2]);
print();

print("Return every other element from the entire array");
print(arr[::2]);
print();
```

```
#--------------------------------------------------Inheritance---------------------------------------------#
#To create a class that inherits the functionality from another class, send the parent class as a parameter when creating the child class:

#Eg:
class Parent:
  x = 10;

class Child(Parent):
  def Print_X(self):
    print(self.x)


myObj = Child()
myObj.Print_X()
```

```
#--------------------------------------------------Inheritance---------------------------------------------#
class Parent:
  x = 10;
  def Print_X(self):
    print("Printed from parent: ", self.x)

class Child(Parent):
  def Print_X(self):
    super().Print_X()
    print("Printed from child: ", self.x)

myObj = Child()
myObj.Print_X()
```
